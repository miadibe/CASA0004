```python
# 1st pretrained 

from keras.applications.vgg16 import VGG16

#Define our model
model = VGG16()
model.summary()
```

```python
#Classify image
from keras.preprocessing.image import load_img
image = load_img('./2006251300008904.jpg', target_size=(224, 224))
```

```python
#Convert image from RGB to a numpy array format
from keras.preprocessing.image import img_to_array
image = img_to_array(image)
```

```python
#Array needs to be in a shape knowing that it's multiple images
#Change the dimensions
#one way of doing it: image.reshape

#another way:
import numpy as np
image = np.expand_dims(image, axis=0)
```

```python
#preprocessing step

from keras.applications.vgg16 import preprocess_input
image = preprocess_input(image)
print(image)
```

```python
#now we're all set and need to predict. 

pred = model.predict(image)
#probability of my image belonging to that specific class is: 
print(pred)
```

```python
#print out the exact top 5 predictions: use tensorflow

from tensorflow.keras.applications.mobilenet import decode_predictions

pred_classes = decode_predictions(pred, top=10)
for i in pred_classes[0]:
    print(i)
    
    #it predicted that there's a crane a 72% !
```
