II. SETUP DATA TABLE 

```python
# Create 2 lists storing respectively all site and BIM filenames 

camera_list = []
for filename in glob.glob('./Train_images-withfloornumber/*.jpg'):
    im = Image.open(filename)
    im = im.filename
    camera_list.append(im)
    
bim_list = []
for filename in glob.glob('./Train_masks-withfloornumber/*.png'):
    im2 = Image.open(filename)
    im2 = im2.filename
    bim_list.append(im2)
    
print(camera_list)
print(bim_list)
```

```python
# Site Images Temporal Data & Number of floors

df_camera = pd.DataFrame(camera_list, columns= ['Camera_Filename'])

df_camera['Date'] = df_camera['Camera_Filename']
df_camera['Camera_Floors'] = df_camera['Camera_Filename']
df_camera['Camera_Filename'] = df_camera['Camera_Filename'].str.replace('./Train_images-withfloornumber/','')
df_camera['Date'] = df_camera['Date'].str.replace('./Train_images-withfloornumber/','')
df_camera['Date'] = df_camera['Date'].str.slice(stop= 6)

df_camera['Camera_Floors'] = df_camera['Camera_Floors'].str.replace('./Train_images-withfloornumber/','')
df_camera['Camera_Floors'] = df_camera['Camera_Floors'].str.slice(start= 17, stop=26)
df_camera['Camera_Floors'] = df_camera['Camera_Floors'].str.replace('.j','')
df_camera['Camera_Floors'] = df_camera['Camera_Floors'].str.replace('.','')

df_camera['CamFloorNum'] = df_camera['Camera_Floors'].str.slice(stop=2)
df_camera['CamFloorNum'] = df_camera['CamFloorNum'].str.replace('-','')
df_camera['CamFloorNum'] = df_camera['CamFloorNum'].astype(int)

df_camera
```

```python
# BIM Images Temporal Data & Number of floors

df_bim = pd.DataFrame(bim_list, columns= ['BIM_Filename'])

df_bim['Date'] = df_bim['BIM_Filename']
df_bim['BIM_Floors'] = df_bim['BIM_Filename']
df_bim['BIM_Filename'] = df_bim['BIM_Filename'].str.replace('./Train_masks-withfloornumber/','')

df_bim['Date'] = df_bim['Date'].str.replace('./Train_masks-withfloornumber/','')
df_bim['Date'] = df_bim['Date'].str.slice(stop=6)

df_bim['BIM_Floors'] = df_bim['BIM_Floors'].str.replace('./Train_masks-withfloornumber/','')
df_bim['BIM_Floors'] = df_bim['BIM_Floors'].str.slice(start= 17, stop=26)
df_bim['BIM_Floors'] = df_bim['BIM_Floors'].str.replace('.p','')
df_bim['BIM_Floors'] = df_bim['BIM_Floors'].str.replace('.','')

df_bim['BIMFloorNum'] = df_bim['BIM_Floors'].str.slice(stop=2)
df_bim['BIMFloorNum'] = df_bim['BIMFloorNum'].str.replace('-','')
df_bim['BIMFloorNum'] = df_bim['BIMFloorNum'].astype(int)

df_bim
```

```python
#Join BIM images with Site images on 'Date'
df = df_camera.merge(df_bim, on='Date')

# Process and extract Date and Time 

df['Year'] = df['Date']
df['Year'] = df['Year'].str.slice(stop=2)
df['Year'] = df['Year'].str.replace('20','2020')
df['Year'] = df['Year'].str.replace('19','2019')
df['Year'] = df['Year'].str.replace('18','2018')

df['Month'] = df['Date']
df['Month'] = df['Month'].str.slice(start=2 , stop=4)

df['Day'] = df['Date']
df['Day'] = df['Day'].str.slice(start=4 , stop=6)

df["DateTime"] = df['Day'] + '-' + df['Month'] + '-'+ df['Year']
df['Temporal'] = pd.to_datetime(df['DateTime'], format='%d/%m/%Y', errors='ignore')

df_time = df.filter(['Temporal','CamFloorNum','BIMFloorNum'])
df_time = df_time.rename(columns={'Temporal':'Date','CamFloorNum':'Site Floors', 'BIMFloorNum': 'BIM Floors'})

df_time
```

```python
FIRST_CAM = './PixelDifference/Train_images'
SECOND_CAM = './PixelDifference/Train_masks'
DIFF_FOLDER = './PixelDifference/Difference'
```

```python
#Compute Pixel difference 

# As a complement, if you want to add one more level of segmentation you can verify if the cv2.subtract has none/any 
# difference with np.all from numpy library and place the files in separate folders

for img_jpg in os.listdir(FIRST_CAM):
    img1 = cv2.imread(os.path.join(FIRST_CAM, img_jpg))
    img_png = img_jpg[:-3] + "png"
    img2 = cv2.imread(os.path.join(SECOND_CAM, img_png))
    difference = cv2.subtract(img1, img2)
    img_diff = 'diff-' + img_png
    # if it doesn't exist, create the DIFF_FOLDER folder.
    diff_level = "NO_DIFF" if np.all(difference == 0) else "SOME_DIFF"
    os.makedirs(DIFF_FOLDER, exist_ok=True)
    cv2.imwrite(os.path.join(DIFF_FOLDER, img_diff), difference)
```
